```python
import http.client
import json
import sys
import time
from datetime import datetime
from argparse import ArgumentParser

data = sys.stdin.read()

archivePort = 28001
loggingPort = 28002


def sendHttp(payload, port):
    conn = http.client.HTTPConnection("localhost", port)
    headers = {
      'Content-Type': 'application/json'
    }
    conn.request("POST", "/", payload, headers)
    res = conn.getresponse()
    data = res.read()
    print(data.decode("utf-8"))

parser = ArgumentParser()
parser.add_argument(
    '-f',
    '--sourceFile',
    dest='sourceFileArg',
    default='')
parser.add_argument(
    '-s',
    '--status',
    dest='statusArg',
    default='success')
parser.add_argument(
    '-d',
    '--loadName',
    dest='loadNameArg',
    default='')
parser.add_argument(
    '-u',
    '--loadUUID',
    dest='loadUUIDArg',
    default='')
parser.add_argument(
    '-ds',
    '--loadStep',
    dest='loadStepArg',
    default='')
parser.add_argument(
    '-m',
    '--message',
    dest='messageArg',
    default='')
parser.add_argument(
    '-a',
    '--action',
    dest='actionArg',
    default='')
parser.add_argument(
    '-r',
    '--records',
    dest='recordsArg',
    default='')

args = parser.parse_args()
sourceFile = args.sourceFileArg
status = args.statusArg
message = args.messageArg
action = args.actionArg
loadName = args.loadNameArg
loadUUID = args.loadUUIDArg
loadStep = args.loadStepArg
records = args.recordsArg

if action == 'log':
  payload = json.dumps({
      "sourceFile": sourceFile,
      "LOAD_DATE_TIME": datetime.today().strftime('%Y-%m-%d %H:%M:%S'),
      "LOAD_NAME": loadName,
      "LOAD_UUID": loadUUID,
      "LOAD_STEP_NAME": loadStep,
      "LOAD_FILENAME": sourceFile,
      "LOAD_STATUS": status,
      "LOAD_RECORDS": records,
      "LOAD_MSG": message
    })
  sendHttp(payload, loggingPort)
elif action == 'archive':
  payload = json.dumps({
      "sourceFile": sourceFile,
      "LOAD_STATUS": status
    })
  sendHttp(payload, archivePort)
```
